```python
inputs = []
outputs = []
with open('input', 'r') as test:
    for line in test:
        inputstr, outputstr = line.strip().split(' | ')
        inputs.append(inputstr.split(' '))
        outputs.append(outputstr.split(' '))
```

```python
import itertools
def part1(inputs, outputs):
    return len(list(filter(lambda x: len(x) in (2, 3, 4, 7), itertools.chain(*outputs))))

part1(inputs, outputs)
```

```python
from tqdm import tqdm
combos = [set(x) for x in ['abcdefg', 'cf', 'acdeg', 'acdfg', 'bcdf', 'abdfg', 'abdefg', 'acf', 'abcdefg', 'abcdfg']]

out_vals = []
for i, o in tqdm(zip(inputs, outputs), total=len(inputs)):
    for perm in itertools.permutations('abcdefg'):
        mapping = {code: decode for code, decode in zip('abcdefg', perm)}
        decoded = [{mapping[l] for l in word} for word in i]
        if all(word in combos for word in decoded):
            # all matches, must be a valid decoding
            decode_outs = [{mapping[l] for l in word} for word in o]
            out_vals.append(int(''.join([str(combos.index(word)) for word in decode_outs])))
```
